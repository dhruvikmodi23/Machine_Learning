```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color = 'green')
```

```python
plt.plot(x,y,color = 'green', linewidth = 5)
```

```python
plt.plot(x,y,color = 'green', linewidth = 5, linestyle ='dashdot')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color = 'green', linewidth = 5, linestyle ='dashdot')
```
