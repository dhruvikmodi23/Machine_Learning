```python
import pandas as pd
import sqlalchemy
```

```python
engine = sqlalchemy.create_engine('mysql+pymysql://root:[pass]@localhost:/application')
```

```python
df = pd.read_sql_table("customers",engine,columns=['name'])
df
```

```python
df = pd.read_sql_query("select id,name from customers",engine)
df
```

```python
query = '''
 SELECT customers.name, customers.phone_number, orders.name, orders.amount
 FROM customers INNER JOIN orders
 ON customers.id=orders.customer_id
'''
df = pd.read_sql_query(query,engine)
df
```

```python
query = '''
 SELECT customers.name, customers.phone_number, orders.name, orders.amount
 FROM customers INNER JOIN orders
 ON customers.id=orders.customer_id
'''
pd.read_sql(query,engine) 
```

```python
df = pd.read_csv("customers.csv")
df
```

```python
df.rename(columns={
    'Customer Name': 'name',
    'Customer Phone': 'phone_number'
}, inplace=True)
df
```

```python
df.to_sql(
    name='customers', # database table name
    con=engine,
    if_exists='append',
    index=False
)
```
