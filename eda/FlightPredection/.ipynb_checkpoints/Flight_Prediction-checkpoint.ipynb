```python
# Flight Price Prediction(EDA + Feature Engineering)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
```

```python
test_df.head()
```

```python
final_df = pd.concat([train_df, test_df], axis=0).reset_index(drop=True)
```

```python
final_df.head()
```

```python
final_df.tail()
```

```python
# feature engineering
```

```python
final_df.info()
```

```python
## date_of_journey -> date , month, year
```

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.drop(['Date_of_Journey'], axis= 1,inplace = True)
```

```python
final_df
```

```python
final_df['Date'] = final_df['Date'].astype(int)
```

```python
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute'] = final_df['Arrival_Minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(['Arrival_Time'],axis =1, inplace = True)
```

```python
final_df.head()
```

```python
final_df['Dep_Hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_Minute'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df.drop(['Dep_Time'],axis =1, inplace = True)
```

```python
final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int)
final_df['Dep_Minute'] = final_df['Dep_Minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Route'].unique()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df['Total_Stops'].isnull()
```

```python
final_df.drop(['Route'],axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.info()
```

```python
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```
